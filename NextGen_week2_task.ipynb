### Importing Data using Pandas read_csv

```python
df=pd.read_csv("Pokemon.csv")
df.head()
```

## Sorting and describing data
### Describing data

```python
df.describe()
```

### Sort name by alphabetical order

```python
df.sort_values("Name").head(10)
```

### Sort by "HP" in descending order

```python
df.sort_values("HP", ascending=False).head(10)
```

### Sort by Attack and Defense in descending order

```python
df.sort_values(["Attack","Defense"],ascending=[0,0]).head(10)
```

## Filtering Data
### Retrive data such that Type1 = Grass (filtering condition Type1 == "Grass")

```python
df.loc[df["Type 1"]=="Grass"].tail(10)
```

### Filtering condition Type1 == "Grass" & Type2 == Poision

```python
df.loc[(df["Type 1"]=="Grass") & (df["Type 2"]=="Fighting")]
```

### Get numerical data columns

```python
numerical_columns=df.select_dtypes(include='number').columns
numerical_columns
```

## Groupby
### Get mean groupby Type 1

```python
df.groupby("Type 1")[df.select_dtypes(include='number').columns].mean()
```

### Get mean groupby Type 2 and sort them by descending "Total"

```python
df.groupby(["Type 2"])[numerical_columns].mean().sort_values(["Total"],ascending=False)
```

### Get count of each Type 1 Pokemon

```python
df["Count"]=1
df.groupby(["Type 1"]).count()["Count"]
```

### Count the number of pokemon with groupby "Type 1" and "Type 2"

```python
df.groupby(["Type 1","Type 2"]).count()["Count"]
```
